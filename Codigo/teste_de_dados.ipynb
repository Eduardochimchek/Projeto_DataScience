```python
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu

dadosLimpos = pd.read_csv('../Data/dados_limpos.csv')
```

```python
# Escolher os desenvolvedores para comparar
dev1 = 'devs_Luiz Gustavo'
dev2 = 'devs_Jorge Andrade'

# Filtrar os dadosLimpos para os desenvolvedores escolhidos
amostra1_sprints = dadosLimpos[dadosLimpos[dev1] == 1]['total_sprints']
amostra1_tempo = dadosLimpos[dadosLimpos[dev1] == 1]['tempo_resolucao']

amostra2_sprints = dadosLimpos[dadosLimpos[dev2] == 1]['total_sprints']
amostra2_tempo = dadosLimpos[dadosLimpos[dev2] == 1]['tempo_resolucao']

# Teste de Mann-Whitney U para amostras independentes nas variáveis de tempo
statistic_tempo, p_value_tempo = mannwhitneyu(amostra1_tempo, amostra2_tempo)

# Teste de Mann-Whitney U para amostras independentes nas variáveis de sprints
statistic_sprints, p_value_sprints = mannwhitneyu(amostra1_sprints, amostra2_sprints)

# Definindo um nível de significância
alpha = 0.05

# Tomando uma decisão com base no valor-p
if p_value_tempo < alpha:
    resultado_teste_tempo = "Rejeitar a hipótese nula - há uma diferença significativa nos tempos de resolução."
else:
    resultado_teste_tempo = "Aceitar a hipótese nula - não há evidências suficientes para afirmar uma diferença significativa nos tempos de resolução."

if p_value_sprints < alpha:
    resultado_teste_sprints = "Rejeitar a hipótese nula - há uma diferença significativa no número total de sprints."
else:
    resultado_teste_sprints = "Aceitar a hipótese nula - não há evidências suficientes para afirmar uma diferença significativa no número total de sprints."

# Plotar gráficos para comparar as amostras
plt.figure(figsize=(12, 6))

# Gráfico de barras para tempos de resolução
plt.subplot(1, 2, 1)
plt.bar([dev1, dev2], [amostra1_tempo.mean(), amostra2_tempo.mean()])
plt.xlabel('Desenvolvedor')
plt.ylabel('Tempo Médio de Resolução')
plt.title('Comparação de Tempos de Resolução entre Desenvolvedores')

# Gráfico de barras para total de sprints
plt.subplot(1, 2, 2)
plt.bar([dev1, dev2], [amostra1_sprints.mean(), amostra2_sprints.mean()])
plt.xlabel('Desenvolvedor')
plt.ylabel('Total Médio de Sprints')
plt.title('Comparação de Total de Sprints entre Desenvolvedores')

plt.tight_layout()
plt.savefig('../Imagens/teste01.jpg')
plt.show()

# Exibir os resultados dos testes
print("Resultado do Teste para Tempos de Resolução:")
print(resultado_teste_tempo)

print("\nResultado do Teste para Total de Sprints:")
print(resultado_teste_sprints)
```
